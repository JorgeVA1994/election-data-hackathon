```python
# Create a CountVectorizer for parsing/counting words
#count_vectorizer = CountVectorizer(ngram_range=(1,2), stop_words='english', token_pattern='\\b[a-z][a-z]+\\b')
count_vectorizer = CountVectorizer(ngram_range=(2,2), stop_words='english', token_pattern='\\b[a-z][a-z]+\\b')
#text = [Clinton2, Trump1, Holt]
text = [Clinton2, Trump1]

#text = Clinton2
count_vectorizer.fit(text)

counts = count_vectorizer.transform(text).transpose()

# Convert sparse matrix of counts to a gensim corpus
corpus = matutils.Sparse2Corpus(counts)

##### Map matrix rows to words (tokens)

id2word = dict((v, k) for k,v in count_vectorizer.vocabulary_.iteritems())

count_vectorizer.vocabulary_.iteritems()

lda = models.LdaModel(corpus=corpus, num_topics=5, id2word=id2word, passes = 10)

lda.print_topics()
```
